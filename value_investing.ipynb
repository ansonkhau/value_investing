```python
import numpy as np
import pandas as pd
import glob 
```

```python
pd.set_option('display.max_columns', 500)
```

```python
DATA_DIR = 'data/'
```

```python
# Get List of Companies
companies = glob.glob(DATA_DIR+"*-income-statement.csv")
for x in xrange(len(companies)):
    companies[x] = companies[x].replace('-income-statement.csv', '')
    companies[x] = companies[x].replace('data/', '')
```

```python
income_state = {}
balance_sheet = {}
key_ratios = {}
pvr_ratios = {}
```

```python
for i in companies:
    income_state[i] = pd.read_csv(DATA_DIR + i +'-income-statement.csv', header = 7, index_col=0, parse_dates='Period End Date')
    balance_sheet[i] = pd.read_csv(DATA_DIR + i +'-balance-sheet.csv', header = 7, index_col=0, parse_dates='Period End Date')
    key_ratios[i] = pd.read_csv(DATA_DIR + i +'-key-ratios.csv', index_col=0, header = 7, parse_dates='Period End Date')
    pvr_ratios[i] = pd.read_csv(DATA_DIR + i +'-pvr-ratios.csv', index_col=0, header = 7, parse_dates='Period End Date')
    income_state[i] = income_state[i].transpose()
    income_state[i] = income_state[i][1:].sort_index(axis=0, by=None, ascending=False)
    income_state[i]['Company'] = i
    income_state[i] = income_state[i].fillna(0)
    balance_sheet[i] = balance_sheet[i].transpose()
    balance_sheet[i] = balance_sheet[i][1:].sort_index(axis=0, by=None, ascending=False)
    balance_sheet[i]['Company'] = i
    key_ratios[i] = key_ratios[i].transpose()
    key_ratios[i] = key_ratios[i][1:].sort_index(axis=0, by=None, ascending=False)
    key_ratios[i]['Company'] = i
    pvr_ratios[i] = pvr_ratios[i].transpose()
    pvr_ratios[i] = pvr_ratios[i][1:].sort_index(axis=0, by=None, ascending=False)
    pvr_ratios[i]['Company'] = i
```
